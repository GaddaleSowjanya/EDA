**how to treat outliers**



**method-1: drop the outliers**
    
- if outliers has less than 2% of total data we can drop the outliers

- droping the outliers is not a recommended approach

- if we drop the outliers other columns data also drop

- we encounter a situation that looses the data

**method-2**

- we already know that ouliers does not affect by median

- so it is always a good approach to fill outliers with median

**method-3**

- instead of filling the outliers with median, we can fill with q3 and q1

```python
# step-1: lb and ub
# step-2: calculate median
# step-3: new data=[]
# step-4: iterate ehc value from prevailing wage
# step-5: if that value <lb or >up
# step-6: new data.append(median)
# step-7: else:
# step-8: new data.append(value)
# step-9: visa_df['pwage']=new data
```

```python
wage_data=visa_df['prevailing_wage']
q1=round(np.quantile(wage_data,0.25),2)
q3=round(np.quantile(wage_data,0.75),2)
IQR=q3-q1
lb=q1-1.5*IQR
ub=q3+1.5*IQR
median=wage_data.median()
new_data=[]
for i in wage_data:
    if i<lb or i>ub:
        new_data.append(median)
    else:
        new_data.append(i)
visa_df['pwage']=new_data
```

**np.where**

- np.where is a mehtod to do if-else task in single line

- np.where(condition,TrueValue,FalseValue)

- it takes 3 arguments

    - condition
    
    - if condition is True then it takes True value
    
    - if conditon is Flase it takes false value

```python
dict1={'marks':[100,200,300,400],
      'sub':['DS','DE','CS','IOT']}
df=pd.DataFrame(dict1)
df
```

```python
# i want to replace marks=1 which are having >200
con=df['marks']>200
true=1
false=df['marks']
df['new']=np.where(con,true,false)
df
```

```python
wage_data=visa_df['prevailing_wage']
q1=round(np.quantile(wage_data,0.25),2)
q3=round(np.quantile(wage_data,0.75),2)
IQR=q3-q1
lb=q1-1.5*IQR
ub=q3+1.5*IQR
median=wage_data.median()
# replace outliers with median using np.where
con=(visa_df['prevailing_wage']<lb) | (visa_df['prevailing_wage']>ub)
true=median
false=visa_df['prevailing_wage']
visa_df['p_wage1']=np.where(con,true,false)
```
