**Data transformations**

- generally data has 3 types of distribution

    - normal distibution
    
    - positive skewed
    
    - negative skewed
    
- all the maths concepts developed by assumption as data follows normal distribution

- data skew happens because of many reasons, in that one reason is outliers

- once we perform the outlier analysis then we can apply transformation methods, if the data does not follows normality

- we have some transformation methods mentioned below

    - log transformation
    
    - exponential transformation
    
    - reciprocal transformation
    
    - square root transformation
    
    - power transformation

$step-1$

```python
import numpy as np
import matplotlib.pyplot as plt
```

$step-2$

```python
exp_data=np.random.exponential(size=10000)
exp_data
```

$step-3$

**plot the data**

```python
plt.hist(exp_data,bins=20,label='exponential')
plt.show()
```

```python
norm_data=np.random.normal(size=10000)
plt.hist(norm_data,bins=20,label='exponential')
plt.show()
```

**goal**

**convert exp_data to normal distribtion**

- bins=interval

$step-4$

**log transformation**

- log transformation means performing operations on original data

- it is an aprocah to convert into normal distribution

- log means natural logarithm base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
log_data
```

```python
plt.hist(log_data)
```

```python
plt.hist(log_data,bins=20,label='logarithm')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label='lograithm')
plt.legend()
plt.show()
```

**reciprocal transformations**

- reciprocal will fail when values are low

- for example: x=0 then 1/0 fail

```python
x=2
1/x
```

```python
reciprocal_data=np.reciprocal(exp_data)
plt.hist(reciprocal_data,bins=20,label='reciprocal')
plt.legend()
plt.show
```

**square root transformation**

```python
square_root_data=np.sqrt(exp_data)
plt.hist(square_root_data,bins=20,label='square root data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,3))
log_data=np.log(exp_data)
plt.subplot(1,3,1).hist(exp_data,bins=20,label='exponential')
plt.legend()
plt.subplot(1,3,2).hist(log_data,bins=20,label='logarithm')
plt.legend()
square_data=np.sqrt(exp_data)
plt.subplot(1,3,3).hist(square_data,bins=20,label='square')
plt.legend()
plt.show()
```

**power transformation**

- box cox

```python
import numpy as np
from sklearn.preprocessing import PowerTransformer
```

```python
from sklearn.preprocessing import PowerTransformer
power=PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)
data_trans=power.fit_transform(exp_data1)
plt.hist(data_trans)
```
