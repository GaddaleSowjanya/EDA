**one hot encoder**

- one hot means at a time only one will **on** and another is **off**

- **on** represent with 1 **off** represent with 0

- on hot encoder creates new columns 

- for example case_status has 2 unique labels

- number of new columns = number of unique labels

- case_status has 2 unique labels one is certified and another one is denied 

- it will create 2 new columns with

    - case_status_Certified
    
    - case_status_Denied
    
- original column as prefix

|case_status|case_status_Certified|case_status_Denied
|-|-|-|
|Certifed|1|0|
|Denied|0|1|
|Denied|0|1|
|certifed|1|0|

**advantages**

- generally ML models expect independance between input variables

- there is relation between variables

- because of one hot at a time only one will be on and another is off

- this indicates the new columns are independent to each other

- this independent behaviour also called as **orthogonality**

- independence means no relation

- no relation means perpendicular each other

- perpendicular means 90 degrees phase shift

- 90 degrees means orthogonal

**disadvantages**

- creates more columns

- more columns mens more data

- more data means more complexity 

- this is called **curse of dimensionality**
