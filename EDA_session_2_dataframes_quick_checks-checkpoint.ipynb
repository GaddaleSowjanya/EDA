**iloc-loc**

- we have drawback with **take**

- take method will give either rows or columns at a time

- but iloc and loc wil give both rows and columns at a time

```python
# syntax

# datframename.iloc[rows,columns]

# datframename.iloc[start:stop:step,start:stop:step]
```

```python
visa_df.iloc[2:5,1:2]
```

```python
visa_df.iloc[200:300,1:2]
```

```python
visa_df.iloc[[100,200,300],[2,6,8]]
```

```python
visa_df.iloc[100:105,[2,6,8]]
```
