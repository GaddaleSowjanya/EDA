```python
income_data=loan_df['ApplicantIncome']
q1=round(np.quantile(income_data,0.25),2)
q3=round(np.quantile(income_data,0.75),2)
IQR=q3-q1
lb=q1-1.5*IQR
ub=q3+1.5*IQR
median=income_data.median()
new_data=[]
for i in income_data:
    if i<lb or i>ub:
        new_data.append(median)
    else:
        new_data.append(i)
loan_df['ApplicantIncome']=new_data
```
