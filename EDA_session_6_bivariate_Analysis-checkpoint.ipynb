```python
cat=visa_df.select_dtypes(include='object').columns
num=visa_df.select_dtypes(exclude='object').columns
```

```python
# consider two columns
# continent and case status
# we already now that continent column has total 6 labels are there
# case status has two labels denied and certified
# how many asia people got certified
# how many asia people got denied
```

```python
# step-1: read the continent
# step-2: read the case status
# step-3: con1= continent=='asia'
# step-4: con2= case staus='certified'
# step-5: con3= con1 and con2
# ste-6: len(visa_df[con3])

con1=visa_df['continent']=='Asia'
con2=visa_df['case_status']=='Certified'
con3=con1&con2
len(visa_df[con3])
```

```python
labels=visa_df['continent'].unique()
for i in labels:
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=='Certified'
    con3=con1&con2
    count=len(visa_df[con3])
    print(f'the number of certified applicants from {i} is: {count}')
```

```python
labels=visa_df['continent'].unique()
certified,denied=[],[]
for i in labels:
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=='Certified'
    con3=visa_df['case_status']=='Denied'
    certi_con=con1&con2
    den_con=con1&con3
    certified.append(len(visa_df[certi_con]))
    denied.append(len(visa_df[den_con]))
certified,denied
```

```python
pd.DataFrame(zip(certified,denied),index=labels,columns=['Certified','Denied'])
```
